#### **The Beginning of Science and Rational Philosophy**

## **Plato's Allegory of the Cave and the Nature of Reality**

**Plato is a central figure in Western philosophy**, along with his teacher Socrates and his most famous student, Aristotle. In a time of crisis for Athens (400 BC), Plato sought a path to absolute truth (Truth with capital T), culminating in his famous **Allegory of the Cave**, contained in the book The Republic.

### **The Allegory of the Cave**
- **Imagine people who have been chained since birth inside a cav**e, unable to move and unable to see anything other than a wall in front of them.
- Behind them, a **fire casts the shadows of statues and objects** held by other people **on the wall**.
- For them, **reality is represented only by these shadows**, because **they know nothing outside of the**m.
- **If** one of these people were **freed and taken out of the cave**, they would **discover a much larger and more complex world**, but **they would have difficulty convincing the other prisoners** that their vision of reality was illusory.

### **Meaning of Metaphor**
Plato uses this allegory to **demonstrate how the senses can deceive us**: **what we perceive** with sight, hearing and touch **is not the true reality**, but **only a reflection of it**.
**True knowledge** does not come from the senses, but **from reason and abstract thought**.
An example: the idea of ​​a perfect circle exists only in the mind; any circle drawn in the real world will never be perfect.

### **Philosophical Influences**
- **Parmenides** → The idea of ​​an immutable essence of reality (Being).
- **Pythagoras** → The belief that mathematics is the key to understanding truth.

## **Aristotle's View of the Cosmos**
Plato did not dwell much on the structure of the cosmos, but his student **Aristotle** developed a detailed theory, which remained the dominant model until the 16th century.

### **1. The Geocentric Model**
- **The Earth at the Center** → Aristotle imagined the cosmos as an **onion**, with the Earth at the center and the known planets (Moon, Mercury, Venus, Sun, Mars, Jupiter, Saturn) orbiting it.
- **The Sphere of Fixed Stars** → Beyond Saturn was the sphere of fixed stars, which looked like crystals set in a celestial dome.
- **The Crystalline Spheres** → The planets were set in transparent spheres that revolved around the Earth, dragging them with them.

### **2. Two Realities and Two Types of Gravity**
Aristotle divided the universe into two regions:
- **Under the Moon (terrestrial world)** → Dominated by the **four elements** (Earth, Water, Air, Fire). Here everything changed and transformed.
- **Above the Moon (celestial world)** → Composed of a fifth element, the **ether** or **quintessence**, eternal and immutable.

This division also influenced gravity:
- **On Earth** → Things fell because their natural movement was to return to their place (e.g. stones fall, fire rises).
- **In the sky** → Celestial bodies moved **in perfect circles**, because circular motion was considered natural and perfect.

### **3. The First Unmoved Mover**
Aristotle addressed the problem of the **first cause**:
- Why does the cosmos move?
- Who gave the first impulse?

His answer was the **Primum Mobile**, an external sphere that transmitted motion to the entire universe. The **unmoved mover**, a divine entity, gave rise to this motion without moving itself.

### **4. The Decline of the Aristotelian Model**
Aristotle's model remained dominant for centuries until, in the 16th century, **Copernicus** introduced the **heliocentric** model, placing the Sun at the center of the universe and challenging the Aristotelian system.

### **The Copernican Revolution and the Decline of the Aristotelian Model**

In **1543**, **Nicolas Copernicus** published *On the Revolutions of the Celestial Spheres*, introducing a revolutionary idea: **the Sun, and not the Earth, is the center of the cosmos**.

### **1. Copernicus' Heliocentric Model**
- Copernicus, influenced by **Renaissance symmetry and proportion**, reorganized the cosmos based on the revolution times of the planets around the Sun:
- **Mercury** → 3 months
- **Venus** → 8 months
- **Earth** → 1 year
- **Mars** → 2 years
- **Jupiter** → 12 years
- **Saturn** → 29 years

- This new vision **directly contradicted Aristotle**. If the Earth was no longer the center, then:
- **Why did objects fall to the ground?**
- **Did humanity still have a special role in the universe?**
- **Were there other beings on Mars or Saturn?**

The problem was that **Copernicus had no new physics to support his theory**. To understand gravity, we had to wait for Galileo, Kepler and Newton.

### **2. Tycho Brahe and the First Denials of Aristotle**
- **Tycho Brahe**, a Danish astronomer with immense financial resources, built **instruments of unprecedented precision** to observe the sky with the naked eye (telescopes did not exist until 1608).
- **1572 - Brahe's Supernova** → A sudden “new star” appeared in the sky.
- Aristotle argued that the heavens could not change, but Brahe proved that this star was located **beyond the Moon**, disproving Aristotle's separation between the terrestrial and celestial worlds.
- **1577 - The Great Comet** → Brahe observed a comet that **passed through the sphere of Mars**, proving that **crystalline spheres did not exist**.

### **3. A Crisis in Cosmology**
After Brahe's discoveries, the Aristotelian model was revealed to be **deeply flawed**:
- The heavens were not **immutable**.
- There were no **crystalline spheres** carrying the planets.

The universe was no longer clear, and **humanity found itself in a period of great uncertainty**.

### **4. The Beginning of the Scientific Revolution**
In **1609**, **Galileo Galilei** pointed the first telescope at the sky and saw a cosmos **completely different from the Aristotelian one**. This marked the real turning point in our understanding of the universe.

**Conclusion**: The discoveries of Copernicus, Brahe and Galileo showed how our view of the cosmos **changes as our observational instruments improve**. This transformation is one of the central themes of the course: **reality is always more complex than we initially perceive.**

### **Free Fall Experiment**

This experiment demonstrates the **effect of air resistance** on free-falling objects.

1. **First Experiment (with air)**
- A **quarter and a piece of paper** of the **same shape are dropped**.
- The **quarter hits the ground first**, while the **paper falls more slowly due to air resistance**.

2. **Second Experiment (in vacuum)**
- A **cylinder connected to a pump is used to remove the air**, creating a **reduced-pressure environment**.
- When the **experiment is repeated**, **both** objects **fall** at exactly the **same speed**.
- This shows that, **in the absence of air**, **mass** does not affect the speed of falling: *all objects accelerate equally under the action of gravity*.

3. **Scientific Implications**
- **Galileo** had already intuited this principle 400 years ago, without the need for an artificial vacuum.
- **The Apollo 15 astronauts** repeated the **experiment on the Moon**, demonstrating that, **without an atmosphere**, objects of **different masses fall at the same speed**.
- Galileo did not limit himself to philosophical speculations, but introduced the experimental method to understand nature with precision.
- His discoveries and astronomical observations challenged the Aristotelian view of the universe, paving the way for a new conception of physical reality.

This experiment teaches us that **reality can be more complex than it appears at first glance**: to understand it, **we must go beyond our perceptions and use science to reveal the laws that govern the world**.

The **formula** that describes these concepts is Galileo's **law of free fall**, based on **gravitational acceleration**:


$$s = \frac{1}{2} g t^2$$


Where:
- $( s )$ is the distance traveled by the falling object (height from which it falls),
- $( g )$ is the gravitational acceleration $(≈ ( 9.81 \, m/s^2 )$ on Earth),
- $( t )$ is the time of fall.

This equation shows that **the distance traveled by a body in free fall is proportional to the square of the time of fall**, regardless of its mass.

Furthermore, another important equation is that of the velocity in free fall:


$$v = g t$$


Where:
- $( v )$ is the velocity of the object after a time $( t )$,
- $( g )$ is always the gravitational acceleration.

This equation indicates that **the velocity of an object in free fall increases linearly with time**, and that it does not depend on the mass of the object.

**If air resistance is negligible (as in a vacuum), all objects fall with the same acceleration $( g )$, regardless of their mass.**

This concept was confirmed in the experiment of the Apollo 15 astronauts on the Moon, where a feather and a hammer hit the ground at the same time.

### **Kepler and Newton's Revolution in Understanding the Cosmos**

While Galileo was revolutionizing the scientific method in Italy, **Johannes Kepler** in Prague was working to understand planetary orbits using data collected by **Tycho Brahe**.

#### **Kepler's Discoveries:**
- For centuries, it had been believed that planets moved in **circular orbits**, which Platonic thought were perfect.
- Analyzing the data, Kepler discovered that **the orbits were elliptical, not circular**.
- This led him to formulate the **three laws of planetary motion**, which for the first time described celestial motion with mathematical precision.
- He also realized that the planets moved around the Sun because there was a **force** that pushed them in that motion, the intensity of which decreased with the distance from the Sun.
- However, Kepler failed to formulate a universal theory of gravity.

#### **Newton's unification:**
- After Kepler, **Isaac Newton** came on the scene, solving the mystery of the force that governs the motion of the planets.
- Newton understood that **the same force that makes an object fall to the Earth is the same that makes the planets orbit around the Sun**.
- Formulates the **Universal Law of Gravitation (1686)**:


$F = SOL \frac{m_1 m_2}{r^2}$


Dove:
- $( F )$ is the gravitational force between two objects,
- $( G )$ is the gravitational constant,
- $( m_1 )$ and $( m_2 )$ are the masses of the two objects,
- $( r )$ is the distance between their centers of mass.

- Newton will show how gravity governs **both the motion of terrestrial bodies and that of celestial objects**, unifying the physics of the Earth and the sky.
- His theory explains phenomena such as:
- **The tides**, caused by the gravitational attraction of the Moon and the Sun.
- **The motion of comets**, such as the 76-year period of Halley's Comet.
- **The motion of the Moon**, describing it as a battle in orbit around the Earth.

#### **The Revolutionary Impact**
- Before Newton, terrestrial and celestial reality were perceived as separate. With his theory, the universe became a **system governed by precise mathematical laws**.
- At the end of the 17th century, this new vision of the cosmos radically changed the way humanity conceived the universe.
- People began to believe that, with the right scientific method, **it would be possible to discover the essence of reality itself**.

---

Newton's intuition marked one of the greatest scientific revolutions: for the first time in history, the universe was no longer an unfathomable mystery, but a **mechanism governed by understandable and predictable physical laws**.

### **[Newton’s General Scholium](https://isaacnewton.ca/newtons-general-scholium/)**

The *General Scholium*, added by Newton to the second edition of the *Principia* in 1713, is one of his most famous and least understood texts. Its complexity comes from the way it is structured: similar to a *matryoshka* (Russian doll), the text hides controversial concepts within more accessible layers.

Modern analysis of Newton’s private manuscripts has allowed scholars to decipher some of Newton’s more cryptic passages, but much work remains to be done.

### **Main Contents of the General Scholium**
- **Critique of Descartes and Leibniz**: Newton refutes Cartesian natural philosophy and responds to Leibniz’s criticisms.
- **Possibility of extrasolar planets**: He hints at the possibility of planets outside our solar system.
- **Rejection of Arbitrary Hypotheses**: Newton argues that science should be based on **induction** and not speculative hypotheses.
- **Speculations on Electricity**: He refers to concepts related to electricity, anticipating future research.
- **Defense of Universal Gravitation**: He reaffirms the importance of the law of universal gravitation.
- **Theology and the Argument from Design**: He discusses the nature of God and uses the argument from *intelligent design* to explain order in the universe.
- **Critique of the Trinity**: Newton includes a veiled criticism of the doctrine of the Trinity, which he believed to be a corruption of Christianity introduced in the 4th century.

### **Newton's General View**
In the *General Scholium*, Newton advocates a **humble and inductive** approach, both in science and religion. He sees natural philosophy and theology as intertwined disciplines, both subject to a process of reform. The text represents a **manifesto of his thought**, in which his vision of the universe emerges as a system governed by precise laws, but also influenced by a divine will.

### **[Can We Ever Prove That the Multiverse Exists?](https://www.smithsonianmag.com/science-nature/can-physicists-ever-prove-multiverse-real-180958813/#3tskFG9MsT7yXIuH.99)**

#### **What is the Multiverse?**
The idea of ​​the multiverse was born from **cosmic inflation**, a very rapid expansion of the universe after the Big Bang. According to the theory of **eternal inflation**, this expansion did not stop everywhere, but continued to create new universes in regions outside our own.

_"Cosmic inflation for dummies: Imagine the infant universe as a buggy code, with errors creating chaos and disorder. Inflation is like a cosmic 'debug': a very rapid expansion that 'stretches' the universe, correcting the errors and making it smooth and flat. According to this theory, the universe was born from a quantum fluctuation, an initial 'error' that triggered the Big Bang, and inflation 'fixed' the code, allowing the universe to 'work' as we see it today."_

_"Eternal Inflation: Imagine the universe as a self-replicating OOP class. The 'false vacuum' is the basic environment, and the 'bubbles' are instances of this class, universes with different physical laws. Chaotic inflation is the cloning method, an infinite loop that creates a multiverse of 'bubbles'. Our universe is just one instance, and the 'fine-tuning' is explained by probability: among infinite universes, one had to be suitable for life."_

There are several interpretations of the multiverse:
1. **Inflationary Multiverse**: the visible universe is only a part of an infinite set of universes born from inflation.
2. **Quantum Multiverse (Many Worlds Interpretation)**: each quantum event with multiple outcomes generates parallel universes.

#### **The Problem of Verifiability**
- The multiverse, by definition, **may be impossible to observe** directly.
- Science requires that theories be **testable** and based on empirical data. Without evidence, the multiverse concept may remain in the **realm of philosophy rather than physics**.
- Some scientists fear that untestable theories will turn physics into a mix of **math, physics and philosophy** with no experimental basis.

#### **Contrasting positions among scientists**
- **Carlo Rovelli and other skeptics**: science must be based on testable data, otherwise it risks returning to medieval thinking.
- **Richard Dawid and supporters**: new definitions of "proof" are needed to evaluate hypotheses that are difficult to test, such as the multiverse.
- **Joseph Polchinski and Andrei Linde**: some features of our universe (its uniformity, the "perfect" physical laws for life) **suggest** the multiverse as a plausible explanation.

#### **Possible future evidence**
- Scientists are looking for **signatures of cosmic inflation** in the cosmic microwave background.
- If our universe had "hit" another in the past, there could be **observable traces** in the cosmos.
- Science has already verified concepts that were initially unprovable (e.g. **gravitational waves** predicted by Einstein and detected after 100 years).

#### **Conclusion**
The multiverse is a fascinating hypothesis, but without direct evidence it remains a speculative theory. Some scientists believe that time and future technologies can confirm it. Others think that, without empirical tests, the multiverse remains a concept more philosophical than scientific.
